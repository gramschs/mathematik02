```python
%matplotlib inline
import matplotlib.pylab as plt
plt.style.use('seaborn')
import numpy as np
```

```python
def periodically_continued(a, b):
    interval = b - a
    return lambda f: lambda x: f((x - a) % interval + a)

x = np.linspace(-5, 5, 1000)

@periodically_continued(0, 3)
def f(x):
    return x**2
```

```python
fig, ax = plt.subplots()
ax.plot(x,f(x))
plt.savefig('mathe02_kap02_blatt03_fig01.png', dpi=300)
```
